```python
import matplotlib.pyplot as plt
%matplotlib inline
def createGraph(fdata ,country):
    fdata_country =fdata[fdata.Country == country]
    plt.plot(fdata_country['Date'], fdata_country['Number of tests per confirmed case'], label=country)
```

```python
createGraph(fdata , "Italy")
createGraph(fdata , "United States")
plt.xlabel("Date")
plt.ylabel("Country")
plt.title("graph1")
plt.legend()
plt.show()
```

```python
fdata_countries = fdata["Country"]
country_list = []
for country in fdata_countries:
    if country not in country_list:
        country_list.append(country)
    else:
        continue
df = pd.DataFrame(country_list , columns= [ "Country"] )
```

```python
countries = ["China" , "United States" , "Italy"]
for i in df.index:
 createGraph(fdata , df["Country"][i])

plt.xlabel("Date")
plt.ylabel("Country")
plt.title("graph1")
plt.legend()
plt.show()
```
